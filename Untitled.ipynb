```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
covid19 = pd.read_csv('owid-covid-data.csv')
```

```python
covid19
```

```python
covid19_latest = covid19[covid19['date']=="2020-05-08"]
```

```python
covid19_latest
```

```python
covid19_latest = covid19_latest.drop(['total_cases_per_million','new_cases_per_million','total_deaths_per_million','new_deaths_per_million','total_tests','new_tests','total_tests_per_thousand','new_tests_per_thousand','tests_units'],axis = 1)
```

```python
covid19_latest.shape
```

```python
covid19_latest = covid19_latest.drop(['iso_code'],axis = 1)
```

```python
covid19_latest.tail()
```

```python
covid19_latest = covid19_latest.drop([15900])
```

```python
covid19_latest =covid19_latest.sort_values(by=['new_deaths'],ascending = False )
plt.figure(figsize = (12,8), dpi =80)
plt.bar(covid19_latest['location'][:12],covid19_latest['new_deaths'][:12],align='center',color='red')
plt.ylabel('Number of Deaths today',size=12)
plt.title('Country with Maximum deaths Latest', size =16)
plt.show()
```

Now we make anlisatiation of individual 12 countries of in graph from thie frist case

```python
covid19_USA =covid19[covid19['location']=='United States']
covid19_USA.reset_index(inplace=True)
covid19_USA.tail()
covid19_USA = covid19_USA.drop(['index','iso_code','total_cases_per_million','new_cases_per_million','total_deaths_per_million','new_deaths_per_million','total_tests','new_tests','total_tests_per_thousand','new_tests_per_thousand','tests_units'],axis =1)
covid19_USA.reset_index(inplace = True)
covid19_USA
covid19_USA.columns = ['Day Count', 'Country', 'Date', 'Total Case', 'New Case', 'Deaths', 'New Deaths' ]
covid19_USA
```

```python
covid19_India = covid19[covid19['location']=='India']
covid19_India.reset_index(inplace = True)
covid19_India
covid19_India = covid19_India.drop(['index','iso_code','total_cases_per_million','new_cases_per_million','total_deaths_per_million','new_deaths_per_million','total_tests','new_tests','total_tests_per_thousand','new_tests_per_thousand','tests_units'],axis =1)
covid19_India.reset_index(inplace = True)
covid19_India
covid19_India.columns = ['Day Count', 'Country', 'Date', 'Total Case', 'New Case', 'Deaths', 'New Deaths']
covid19_India
```

```python
covid19_Brazil = covid19[covid19['location']=='Brazil']
covid19_Brazil.reset_index(inplace = True)
covid19_Brazil
covid19_Brazil = covid19_Brazil.drop(['index','iso_code','total_cases_per_million','new_cases_per_million','total_deaths_per_million','new_deaths_per_million','total_tests','new_tests','total_tests_per_thousand','new_tests_per_thousand','tests_units'],axis =1)
covid19_Brazil.reset_index(inplace = True)
covid19_Brazil
covid19_Brazil.columns = ['Day Count', 'Country', 'Date', 'Total Case', 'New Case', 'Deaths', 'New Deaths']
covid19_Brazil
```

```python
covid19_United_Kingdom = covid19[covid19['location']=='United Kingdom']
covid19_United_Kingdom.reset_index(inplace = True)
covid19_United_Kingdom
covid19_United_Kingdom = covid19_United_Kingdom.drop(['index','iso_code','total_cases_per_million','new_cases_per_million','total_deaths_per_million','new_deaths_per_million','total_tests','new_tests','total_tests_per_thousand','new_tests_per_thousand','tests_units'],axis =1)
covid19_United_Kingdom.reset_index(inplace = True)
covid19_United_Kingdom
covid19_United_Kingdom.columns = ['Day Count', 'Country', 'Date', 'Total Case', 'New Case', 'Deaths', 'New Deaths']
covid19_United_Kingdom
```

```python
plt.figure(figsize=(12,8),dpi=80)
plt.plot(covid19_USA['Day Count'], covid19_USA['New Case'])
plt.plot(covid19_India['Day Count'], covid19_India['New Case'])
plt.plot(covid19_Brazil['Day Count'], covid19_Brazil['New Case'])
plt.plot(covid19_United_Kingdom['Day Count'], covid19_United_Kingdom['New Case'])
plt.legend(['United States','India','Brazil','United_Kingdom'])
plt.xlabel('Day count', size=12)
plt.ylabel('New Cases', size=12)
plt.title('Which Country have Flatting curve?', size = 16)
plt.show()
```
