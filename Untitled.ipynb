```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
covid19 = pd.read_csv('owid-covid-data.csv')
```

```python
covid19
```

```python
covid19_latest = covid19[covid19['date']=="2020-05-08"]
```

```python
covid19_latest
```

```python
covid19_latest = covid19_latest.drop(['total_cases_per_million','new_cases_per_million','total_deaths_per_million','new_deaths_per_million','total_tests','new_tests','total_tests_per_thousand','new_tests_per_thousand','tests_units'],axis = 1)
```

```python
covid19_latest.shape
```

```python
covid19_latest = covid19_latest.drop(['iso_code'],axis = 1)
```

```python
covid19_latest.tail()
```

```python
covid19_latest =covid19_latest.sort_values(by=['new_deaths'],ascending = False )
plt.figure(figsize = (12,8), dpi =80)
plt.bar(covid19_latest['location'][:5],covid19_latest['new_deaths'][:5],align='center',color='blue')
plt.ylabel('Number of Deaths today',size=12)
plt.title('States with Maximum deaths Latest', size =16)
plt.show()
```
